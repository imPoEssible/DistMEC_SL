```python
import numpy as np
import matplotlib.pyplot as plt

from classes.User import *
from classes.solver import *
from classes.Server import *
from classes.distributed_utils import *

import pickle
import copy
```

```python
# Load pre-saved data to analyze
with open('r1_comparisons.p', 'rb') as handle:
    b = pickle.load(handle)
```

```python
# For each trial, calculate relevant values
plot_idx = 6
relevant_dict = b[plot_idx]
num_users = 10

pm_single = 0.05
pm = 1 - (1-pm_single)**(num_users)
po = 0.1

reg_threshold = 0.9
regret_temp = copy.deepcopy(b[plot_idx]['regret_res'])
regret_temp[regret_temp <= 0] = 10
regret_temp[regret_temp <= reg_threshold] = 10

delta_min = min(regret_temp)

constant = delta_min/(po/pm + 1)
lower_bound = np.arange(T)*constant

plt.plot(np.cumsum(b[plot_idx]['regret_res']), label = 'reserve')
plt.plot(np.cumsum(b[plot_idx]['regret_w']), label = 'worst case')
plt.plot(np.cumsum(b[plot_idx]['regret_s']), label = 'new rsv')
plt.plot(lower_bound, label = 'lower bound')

plt.title('Cumulative Regret CMAB')
plt.xlabel('Rounds')
plt.ylabel('Regret')
plt.legend()
```
